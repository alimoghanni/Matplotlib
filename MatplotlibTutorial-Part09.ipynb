```python
# Preamble: useful toolboxes, librairies, functions, etc.

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# load data form csv file and plot

data = pd.read_csv('datamatplotlib-9.csv')
ages = data['Age']
Bentley_y = data['Bentley']
Audi_y = data['Audi']
Cadillac_y = data['Cadillac']

plt.plot(ages, Audi_y, label='Audi')
plt.plot(ages, Cadillac_y, label='Cadillac')

plt.plot(ages, Bentley_y, color='#444444',
         linestyle='--', label='Bentley')

plt.legend()

plt.xlabel('Ages')
plt.ylabel('Number of Different cars User')
plt.title('Number of Different cars User by Age')

plt.tight_layout()

plt.show()
```

```python
# plot in subplots

fig, ax = plt.subplots()

ax.plot(ages, Audi_y, label='Audi')
ax.plot(ages, Cadillac_y, label='Cadillac')

ax.plot(ages, Bentley_y, color='#444444',
         linestyle='--', label='Bentley')

ax.legend()

ax.set_xlabel('Ages')
ax.set_ylabel('Number of Different cars User')
ax.set_title('Number of Different cars User by Age')

plt.tight_layout()

plt.show()
```

```python
# plot is subplot and separate plots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(ages, Bentley_y, color='#444444',
         linestyle='--', label='Bentley')

ax2.plot(ages, Audi_y, label='Audi')
ax2.plot(ages, Cadillac_y, label='Cadillac')

ax1.legend()

ax1.set_xlabel('Ages')
ax1.set_ylabel('Number of Different cars User')
ax1.set_title('Number of Different cars User by Age')

ax2.legend()

ax2.set_xlabel('Ages')
ax2.set_ylabel('Number of Different cars User')
ax2.set_title('Number of Different cars User by Age')


plt.tight_layout()

plt.show()
```

```python
# add custom changes

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(ages, Bentley_y, color='#444444',
         linestyle='--', label='Bentley')

ax2.plot(ages, Audi_y, label='Audi')
ax2.plot(ages, Cadillac_y, label='Cadillac')

ax1.legend()

ax1.set_ylabel('Number of Cars  User')
ax1.set_title('Number of Different cars User by Age')

ax2.legend()

ax2.set_xlabel('Ages')
ax2.set_ylabel('Number of Cars User')

plt.tight_layout()

plt.show()
```

```python
# share x axis

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, Bentley_y, color='#444444',
         linestyle='--', label='Bentley')

ax2.plot(ages, Audi_y, label='Audi')
ax2.plot(ages, Cadillac_y, label='Cadillac')

ax1.legend()

ax1.set_ylabel('Number of Cars User')
ax1.set_title('Number of Different cars User by Age')

ax2.legend()

ax2.set_xlabel('Ages')
ax2.set_ylabel('Number of Cars User')

plt.tight_layout()

plt.show()
```

```python
# separate two figuare in subplots

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, Bentley_y, color='#444444',
         linestyle='--', label='Bentley')

ax2.plot(ages, Audi_y, label='Audi')
ax2.plot(ages, Cadillac_y, label='Cadillac')

ax1.legend()

ax1.set_title('Number of Different cars User by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Number of Cars User')

ax2.legend()

ax2.set_title('Number of Different cars User by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Number of Cars User')

plt.tight_layout()

plt.show()
```
