```python
# Preamble: useful toolboxes, librairies, functions, etc.

import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

### Pie Charts

```python
# simple pie chart

slices = [65, 35]

plt.pie(slices)

plt.title('My First Pie Chart')
plt.tight_layout()
plt.show()
```

```python
# add labels to pie chart

slices = [65, 35]
labels = ['Bentley', 'Audi']

plt.pie(slices, labels=labels)

plt.title('My First Pie Chart')
plt.tight_layout()
plt.show()
```

```python
# add separaters to pie chart

slices = [65, 35]
labels = ['Bentley', 'Audi']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})

plt.title('My First Pie Chart')
plt.tight_layout()
plt.show()
```

```python
# add your colors to pie chart

slices = [65, 35, 20, 14]
labels = ['Bentley', 'Audi', 'Cadillac', 'Kia']
colors = ['red', 'green', 'blue', 'yellow']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'black'})

plt.title('My First Pie Chart')
plt.tight_layout()
plt.show()
```

```python
# use hex colors to pie chart

slices = [65, 35, 20, 14]
labels = ['Bentley', 'Audi', 'Cadillac', 'Kia']
colors = ['#eb3434', '#34eb83', '#34bdeb', '#deeb34']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'black'})

plt.title('My First Pie Chart')
plt.tight_layout()
plt.show()
```

```python
# pie chart with more pies

cars = ['Chevrolet', 'Bentley', 'Ferrari', 'Fiat', 'Audi', 'Cadillac', 'Hyundai',
'Isuzu', 'Jaguar', 'Kia', 'Lamborghini', 'Land Rover', 'Maserati', 'Mazda', 'Peykan']

popularity = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]


plt.pie(popularity, labels=cars, wedgeprops={'edgecolor':'black'})

plt.title('Different cars Use')
plt.tight_layout()
plt.show()
```

```python
# pie chart with 5 pies

cars = ['Peykan', 'Bentley', 'Ferrari', 'Maserati', 'Audi']
popularity = [59219, 55466, 47544, 36443, 35917]


plt.pie(popularity, labels=cars, wedgeprops={'edgecolor':'black'})

plt.title('Different cars Use')
plt.tight_layout()
plt.show()
```

```python
# emphasize on one pie

cars = ['Peykan', 'Bentley', 'Ferrari', 'Maserati', 'Audi']
popularity = [59219, 55466, 47544, 36443, 35917]
explode = [0,0.1,0,0,0]


plt.pie(popularity, labels=cars, explode=explode, wedgeprops={'edgecolor':'black'})

plt.title('Different cars Use')
plt.tight_layout()
plt.show()
```

```python
# add shadow to pie charts

cars = ['Peykan', 'Bentley', 'Ferrari', 'Maserati', 'Audi']
popularity = [59219, 55466, 47544, 36443, 35917]
explode = [0,0.1,0,0,0]


plt.pie(popularity, labels=cars, explode=explode, shadow=True,
        wedgeprops={'edgecolor':'black'})

plt.title('Different cars Use')
plt.tight_layout()
plt.show()
```

```python
# add starting angle pie charts

cars = ['Peykan', 'Bentley', 'Ferrari', 'Maserati', 'Audi']
popularity = [59219, 55466, 47544, 36443, 35917]
explode = [0,0.1,0,0,0]


plt.pie(popularity, labels=cars, explode=explode, shadow=True,
        startangle=90,
        wedgeprops={'edgecolor':'black'})

plt.title('Different cars Use')
plt.tight_layout()
plt.show()
```

```python
# display percentage to pie charts

cars = ['Peykan', 'Bentley', 'Ferrari', 'Maserati', 'Audi']
popularity = [59219, 55466, 47544, 36443, 35917]
explode = [0,0.1,0,0,0]


plt.pie(popularity, labels=cars, explode=explode, shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor':'black'})

plt.title('Different cars Use')
plt.tight_layout()
plt.show()
```

```python
# using style 'fivethirtyeight' 

plt.style.use('fivethirtyeight')

cars = ['Peykan', 'Bentley', 'Ferrari', 'Maserati', 'Audi']
popularity = [59219, 55466, 47544, 36443, 35917]
explode = [0,0.1,0,0,0]


plt.pie(popularity, labels=cars, explode=explode, shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor':'black'})

plt.title('Different cars Use')
plt.tight_layout()
plt.show()
```
