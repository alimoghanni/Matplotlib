```python
# Preamble: useful toolboxes, librairies, functions, etc.

import pandas as pd
import numpy as np

%matplotlib inline

from matplotlib import pyplot as plt
```

```python
# simple scatter plot

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]


plt.scatter(x, y)

plt.tight_layout()

plt.show()
```

```python
# change size of dots

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]


plt.scatter(x, y, s=100)

plt.tight_layout()

plt.show()
```

```python
# change color of dots

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]


plt.scatter(x, y, s=100, c='#29d657')

plt.tight_layout()

plt.show()
```

```python
# change marker of dots

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]


plt.scatter(x, y, s=100, c='#29d657', marker='>')

plt.tight_layout()

plt.show()
```

```python
# add edge and linewidth


x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]


plt.scatter(x, y, s=100, c='#29d657', edgecolor='black', linewidth=1, alpha=0.75 )

plt.tight_layout()

plt.show()
```

```python
# add colors to each dots

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, edgecolor='black', linewidth=1, alpha=0.75 )

plt.tight_layout()

plt.show()
```

```python
# add colors maps


x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, cmap='Oranges', edgecolor='black', linewidth=1, alpha=0.75 )

plt.tight_layout()

plt.show()
```

```python
# add colorbar legend

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, cmap='Oranges', edgecolor='black', linewidth=1, alpha=0.75 )

cbar= plt.colorbar()
cbar.set_label('satisfaction')

plt.tight_layout()

plt.show()
```

```python
# add sizes to dots 

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=sizes, c=colors, cmap='Oranges', edgecolor='black', linewidth=1, alpha=0.75 )

cbar= plt.colorbar()
cbar.set_label('satisfaction')

plt.tight_layout()

plt.show()
```

```python
# load read data and plot scatter plot

data = pd.read_csv('datamatplotlib-7.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75 )


plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

```python
# use log scales

data = pd.read_csv('datamatplotlib-7.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75 )

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

```python
# use ratio and add colormap

data = pd.read_csv('datamatplotlib-7.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes,
            c=ratio, cmap='spring',
            edgecolor='black', linewidth=1, alpha=0.75 )

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

```python
# using style 'seaborn' 

plt.style.use('seaborn')

data = pd.read_csv('datamatplotlib-7.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes,
            c=ratio, cmap='spring',
            edgecolor='black', linewidth=1, alpha=0.75 )

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
