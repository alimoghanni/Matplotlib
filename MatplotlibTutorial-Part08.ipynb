```python
# Preamble: useful toolboxes, librairies, functions, etc.

import pandas as pd
import numpy as np
from datetime import datetime, timedelta

%matplotlib inline

from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
# simple time series plot

dates = [
    datetime(2018, 7, 24),
    datetime(2018, 7, 25),
    datetime(2018, 7, 26),
    datetime(2018, 7, 27),
    datetime(2018, 7, 28),
    datetime(2018, 7, 29),
    datetime(2018, 7, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates,y)

plt.tight_layout()

plt.show()
```

```python
# change line style

dates = [
    datetime(2018, 7, 24),
    datetime(2018, 7, 25),
    datetime(2018, 7, 26),
    datetime(2018, 7, 27),
    datetime(2018, 7, 28),
    datetime(2018, 7, 29),
    datetime(2018, 7, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')

plt.tight_layout()

plt.show()
```

```python
# add auto format to x axis

dates = [
    datetime(2018, 7, 24),
    datetime(2018, 7, 25),
    datetime(2018, 7, 26),
    datetime(2018, 7, 27),
    datetime(2018, 7, 28),
    datetime(2018, 7, 29),
    datetime(2018, 7, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.tight_layout()

plt.show()
```

```python
# change the format of dates on x axis

dates = [
    datetime(2018, 7, 24),
    datetime(2018, 7, 25),
    datetime(2018, 7, 26),
    datetime(2018, 7, 27),
    datetime(2018, 7, 28),
    datetime(2018, 7, 29),
    datetime(2018, 7, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b, %d %Y')

plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()

plt.show()
```

```python
# load data and plot time series

data = pd.read_csv('datamatplotlib-8.csv')

data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
